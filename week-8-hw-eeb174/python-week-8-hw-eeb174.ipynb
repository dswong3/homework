```python
#1st func

def canid_dict(csv):
    from collections import defaultdict
    
    with open(csv, "r") as aa:
        csv_data = aa.readlines()
    
    
    species_ranges = defaultdict(list)


    for line in csv_data:
        line = line.replace('"', '')
        species = line.split(",")[5]
        if len(species.split(" ")) > 1:
            minage = float(line.split(',')[10])
            maxage = float(line.split(',')[11])
            meanage = (minage + maxage) / 2
    
            species_ranges[species].append(meanage) 
    return species_ranges
```

```python
new_dict = canid_dict("sorted_canids.csv")
new_dict
```

```python
#2nd func
def summarized_canid(species_dict, outputfilename):
    output = open(str(outputfilename), "w")
    for key in sorted(species_dict.keys()): 
        ages = species_dict[key]
        maxage = max(ages)
        minage = min(ages)
        genus = key.split(" ")[0] 
        species =key.replace(" ", "_") 
        outline = "{},{},{},{}\n".format(genus, species, minage, maxage)
        output.write(outline)
    output.close()
```

```python
summarized_canid(new_dict, "/home/eeb177-student/Desktop/eeb-177/class-assignments/classwork-2-March/canid-ranges.csv")
```
